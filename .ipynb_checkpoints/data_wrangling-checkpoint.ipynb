```python
import pandas as pd
```

```python
zomato = pd.read_csv("modified_zomato.csv", encoding='latin1')
cuisines = zomato['cuisines'].str.get_dummies(',')

cuisine_count = pd.concat([zomato[['listed_in(city)', 'name']], cuisines], axis = 1)
cuisine_count = cuisine_count.groupby(['listed_in(city)', 'name']).max().reset_index()
final_count = cuisine_count.groupby('listed_in(city)')[cuisine_count.columns[2:]].sum().reset_index()
print(final_count)
```

```python
most_popular = final_count.iloc[:, 1:].idxmax(axis=1)
most_popular = pd.DataFrame({'locality':final_count['listed_in(city)'], 'mp_cuisine': most_popular[0]})
print(most_popular)
```
