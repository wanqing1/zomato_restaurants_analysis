```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
import statsmodels.api as sm
```

### Rate and Cost

The three numerical variables in our data frame are 'cost', 'rate' and 'votes'. Both of 'rate' and 'votes'
could be considered as popularity indicators. From the line chart we draw in Tableau, there seems to be a weak linear relationship between average 'rate' and 'cost', and no clear relationship between average  'votes' and 'cost'. Here we would like to confirm if there is a linear relationship between 'rate' and 'cost'.

```python
zomato = pd.read_csv("cleaned.csv")
```

```python
# Make a scatterplot of rate and cost
plt.figure(figsize=(4, 3))
plt.title('Rate and Cost')
plt.xlabel('Cost')
plt.ylabel('Rate')
plt.scatter(zomato['cost'], zomato['rate'])
```

Probably the relationship is better visualized with side-by-side boxplots by separating the 'cost' into several intervals?

```python
# Separate the cost into intervals.
cost_intervals = [0, 1000, 2000, 3000, 4000, 5000, 6000]
zomato['cost_intervals'] = pd.cut(zomato['cost'], 
                                  bins=cost_intervals, 
                                  labels=['0-1000', '1001-2000', '2001-3000', '3001-4000', '4001-5000', '5001-6000'])
plt.figure(figsize=(5,4))
plt.xticks(rotation=45)
plt.title('Rate vs Cost Intervals')
plt.xlabel('Cost Intervals')
plt.ylabel('Rate')
plt.grid(axis='y')
sns.boxplot(x='cost_intervals', y='rate', data=zomato)
```

Now it's much clear to see the general trend that as the 'cost' increases, 'rate' increases. To figure out if there's a relationship between them, we set up a hypothesis test. The null hypothesis is that there is no relationship between 'rate' and 'cost', while the alternative hypothesis states that there is some relationship between the 'rate' and 'cost'.

```python
# Remove null values, then find the correlation coefficient and p value
zomato_dropna = zomato.dropna(subset=['rate', 'cost'])
rate = zomato_dropna['rate']
cost = zomato_dropna['cost']
corr_coef, p_value = pearsonr(rate, cost)
print(corr_coef)
print(p_value)
```

The correlation coefficient between the two variables is about 0.34 and the p value of our hypothesis testing is close to 0, suggesting that there's a weak linear relationship between 'rate' and 'cost'. Then, we use the least-square method to set up a linear equation that approximates the relationship between 'rate' and 'cost'.

```python
# Use least-square method to build a linear relationship
cost_with_const = sm.add_constant(cost)
model = sm.OLS(rate, cost_with_const).fit()
slope = model.params['cost']
intercept = model.params['const']
print(f"rate = {slope:.5f} * cost + {intercept:.2f}")
```
